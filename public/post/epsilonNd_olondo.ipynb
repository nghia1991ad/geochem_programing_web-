### Calculations of $\epsilon _{Nd}$

$$\epsilon _{Nd} = \left(\frac{(\frac{^{143}Nd}{^{144}Nd})_i^{sample}}{(\frac{^{143}Nd}{^{144}Nd})_i^{CHUR}} -1\right)\times 10^4$$

where :

* $(\frac{^{143}Nd}{^{144}Nd})_t = (\frac{^{143}Nd}{^{144}Nd})_0 - (\frac{^{147}Sm}{^{144}Nd})_0(e^{\lambda t}-1)$ : we need to determine this ratio for both sample and CHUR at time t


* $\lambda_{147_{Sm}} = 6.54\times 10^{-12}y^{-1}$ (Lugmair and Marti 1978)


* $\frac{^{147}Sm}{^{144}Nd} = (\frac{C_{Sm}}{C_{Nd}})\left[0.53151 + 0.14252 \frac{^{143}Nd}{^{144}Nd}  \right]$


* $^{143}Nd/^{144}Nd_{CHUR} = 0.512638$ and $^{147}Sm/^{144}Nd_{CHUR} = 0.1967$ (Jacobsen and Wasserburg 1980)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("olondo_sm_nd.csv")
df.head()
```

```python
df["147Sm/144Nd_re"] = df["Sm"]/df["Nd"]*(0.53151+0.14252*df["143Nd/144Nd"])
Nd143_Nd144CHUR = 0.512638
Sm147_Nd144CHUR = 0.1967
lda = 6.54e-12
df["eNd(0)_re"] = (df["143Nd/144Nd"]/Nd143_Nd144CHUR-1) * 10**4
df["eNd(0)_re"] = df["eNd(0)_re"].round(1)
df["143Nd/144Ndt"] = df["143Nd/144Nd"] - df["147Sm/144Nd_re"]*(np.exp(lda*df["Age"]*10**6)-1)
df["143Nd/144Nd_CHURt"] = Nd143_Nd144CHUR - Sm147_Nd144CHUR*(np.exp(lda*df["Age"]*10**6)-1)
df["eNd(t)_re"] = (df["143Nd/144Ndt"]/df["143Nd/144Nd_CHURt"]-1) * 10**4
df["eNd(t)_re"] = df["eNd(t)_re"].round(2)
```
